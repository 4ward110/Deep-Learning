```python
# ung dung matplot lib de ve do thi

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
print(x)
y_sin = np.sin(x)
print(y_sin)
y_cos = np.cos(x)
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.show()
```
